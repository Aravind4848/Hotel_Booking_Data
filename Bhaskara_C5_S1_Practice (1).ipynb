```python
# defining a function for categorical columns
def info_of_cat(col):
    print(f"Unique values in {col} are: {df[col].nunique()}")
    print(f"Mode values of the {col} are: {df[col].mode()}")
    print(f"Missing values of {col} are: {df[col].isnull().sum()}")
```

```python
# Calculating the unique values , mode and missing values of Attrition
info_of_cat('Attrition')
```

```python
# Calculating the unique values , mode and missing values of Business Travel
info_of_cat('BusinessTravel')
```

```python
# Calculating the unique values, mode, and missing values of Department
info_of_cat('Department')
```

```python
# Filling the missing values of business travel with the mode value i.e., Travel Rarely.
df['BusinessTravel'] = df['BusinessTravel'].fillna('Travel_Rarely')

# Filling the missing department values with the mode value i.e., Research & Development.
df['Department'] = df['Department'].fillna('Research & Development')

# Filling the missing values of the Education field with the mode value i.e., Life Sciences.
df['EducationField'] = df['EducationField'].fillna('Life Sciences')

# Filling the missing values of Gender with the mode value i.e., Male.
df['Gender'] = df['Gender'].fillna('Male')

# Filling the missing values of Gender with the mode value i.e., Male.
df['MaritalStatus'] = df['MaritalStatus'].fillna('Married')

# Filling the missing values of Job Role with the mode value i.e., Sales Executive.
df['JobRole'] = df['JobRole'].fillna('Sales Executive')

# Filling the missing values of Marital Status with the mode value i.e., Married.
df['JobRole'] = df['JobRole'].fillna('Married')

# Filling the missing values of Over18 with the mode value i.e., Y.
df['Over18'] = df['Over18'].fillna('Y')
```
